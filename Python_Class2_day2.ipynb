```python
# pip install requests 원래 깔아야하지만 코랩에 깔아져 있음
import requests # 웹연결 라이브러리
```

```python
from bs4 import BeautifulSoup as bs # 데이터를 분석하기 쉽게 정제 라이브러리
```

```python
import pandas as pd   # 데이터 분석 라이브러리
```

```python
url = 'https://music.bugs.co.kr/chart'
html = requests.get(url)
soup = bs(html.text, 'html.parser') # 글자 그자체를 html로 바꾸는 과정
soup
```

```python
soup.select('tr')
soup.select('tbody > tr')
soup.select('table > tbody > tr')
len(soup.select('table.byChart > tbody > tr'))
```

```python
songs = soup.select('table.byChart > tbody > tr')
```

```python
song = songs[0]  # 첫번째곡의 코드
song
```

```python
singer = song.select(' p.artist > a')[0].text
singer
```

```python
title = song.select('p.title > a')[0].text
title
```

```python
song_data = []
rank = 1

songs = soup.select('table.byChart > tbody > tr')

for song in songs:
  title = song.select('p.title > a')[0].text
  singer = song.select('p.artist > a ')[0].text
  song_data.append([rank,title,singer])
  rank += 1 

song_data
```

```python
df = pd.DataFrame(song_data,columns = ['순위','타이틀','가수'])
df
```

```python
df.to_excel('bugs.xlsx',index=False)
```
